```r
#Alt way to look up snps: 
#/publicdata/1kg_snps
#/data/UKBB_Neale_v2/variants.tsv.gz

get_rsids = function(vect){
rsids = getBM(attributes = c('refsnp_id', 'chr_name', 'chrom_start', 'chrom_end', 'allele'),
      filters = c('chromosomal_region'), 
      values = vect, 
      mart = snpMart) 
    return(rsids)
    }
```

```r
chunk2 <- function(x,n) split(x, cut(seq_along(x), n, labels = FALSE)) 
```

```r
ncores = 32
```

```r
outdir = '/data/variant_lists/'
```

```r
test1 = chunk2(coords,500)
```

```r
length(test1)
```

```r
df2 = df
df2[,1:3] <- sapply(df2[,1:3],as.integer)
```

```r
for (i in 1:500){
test2 = chunk2(test1[[i]],50)
    RS1 = mclapply(test2, get_rsids, mc.cores = ncores )
    tab =rbindlist(RS1[lapply(RS1, class) == "data.frame"])
    m = merge(tab, df2, by=c('chr_name','chrom_start','chrom_end','allele'))
    write.table(m, paste0(outdir, "rsID_part", i, ".txt"))
    }
```
